# Problem 1
Bir sayının 7'ye bölünebilir olup olmadığını kontrol etmek 
için bir program yazın.

İpucu: Mod operatörünü (%) kullanabilirsiniz.

````python
sayi = int(input("Bir sayı giriniz : "))
if sayi % 7 == 0:  #7 ye tam bölünüyor mu ?
    print(f"{sayi} sayısı 7'ye tam bölünüyor")
else:
    print(f"{sayi} sayısı 7'ye tam bölünmüyor")
````

```
 BKİ 18.5'un altındaysa -------> Zayıf
```

# Problem 7
Bir yılın artık yıl olup olmadığını kontrol etmek için bir program yazın. 
Artık yıl kuralı:

a) Eğer yılın son iki basamağı (00) ile bitmiyor ve yalnızca 4'e tam bölünüyorsa artık yıldır. Ya da, 

b) Eğer yıl 400'e tam bölünüyorsa artık yıldır.

**Örneğin**
```
2017 artık yıl değildir.
1900 artık yıl değildir
2012 artık yıldır.
2000 artık yıldır.
```

````python
yıl = int(input("Yılı giriniz: "))

if (yıl % 4 == 0 and yıl % 100 != 0) or (yıl % 400 == 0):
    print(f"{yıl} artık yıldır.")
else:
    print(f"{yıl} artık yıl değildir.")
````
