# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
class UserData:
    def __init__(self, name, surname, age, height, weight):
        self.name = name
        self.surname = surname
        self.age = age
        self.height = height
        self.weight = weight
        self.bmi = self.calculate_bmi()
        self.category = self.bmi_category()

    def calculate_bmi(self):
        return self.weight / (self.height ** 2)

    def bmi_category(self):
        bmi = self.bmi
        if bmi < 18.5:
            return "Zayıf"
        elif 18.5 <= bmi < 24.9:
            return "Normal kilolu"
        elif 25 <= bmi < 29.9:
            return "Fazla kilolu"
        elif 30 <= bmi < 39.9:
            return "Obez"
        else:
            return "İleri derecede obez (morbid obez)"

def get_user_data():
    name = input("İsim: ")
    surname = input("Soyisim: ")
    age = int(input("Yaş: "))
    height = float(input("Boy (metre cinsinden, örn: 1.75): "))
    weight = float(input("Kilo (kg cinsinden): "))
    return UserData(name, surname, age, height, weight)

def main():
    num_entries = int(input("Kaç giriş yapmak istiyorsunuz? "))
    all_users = []

    for _ in range(num_entries):
        user = get_user_data()
        all_users.append(user)

        print(f"İsim: {user.name} {user.surname}, Yaş: {user.age}, Boy: {user.height}m, Kilo: {user.weight}kg")
        print(f"VKİ: {user.bmi:.2f}, Kategori: {user.category}\n")

    # Save data to file
    with open("kullanici_verileri.txt", "w", encoding="utf-8") as file:
        for user in all_users:
            file.write(f"İsim: {user.name}, Soyisim: {user.surname}, Yaş: {user.age}, Boy: {user.height}m, Kilo: {user.weight}kg, VKİ: {user.bmi:.2f}, Kategori: {user.category}\n")

    print("Tüm veriler 'kullanici_verileri.txt' dosyasına kaydedildi.")

if __name__ == "__main__":
    main()

```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import datetime
MENU = {
    "Karpuz Smoothie": 5.00,
    "Mango Smoothie": 6.00,
    "Yaban Mersini Smoothie": 7.00,
    "Yeşil Smoothie": 8.00,
    "Ananas Smoothie": 6.50
}

def display_menu():
    """Menüyü görüntüleyen fonksiyon"""
    print("\nSmoothie Menü:")
    for item, price in MENU.items():
        print(f"{item}: ${price:.2f}")
    print()

def purchase_smoothie():
    """Smoothie satın alma fonksiyonu"""
    display_menu()
    choice = input("Lütfen satın almak istediğiniz smoothie'yi seçiniz: ")
    
    if choice in MENU:
        price = MENU[choice]
        print(f"{choice} seçildi. Toplam fiyat: ${price:.2f}")
        with open("satinalma_kayitlari.txt", "a", encoding="utf-8") as file:
            file.write(f"{datetime.datetime.now()}: {choice} - ${price:.2f}\n")
        
        print("Satın alma işlemi tamamlandı.")
    else:
        print("Geçersiz seçim. Lütfen menüdeki seçeneklerden birini seçiniz.")

def main():
    """Ana program döngüsü"""
    while True:
        print("1. Menüyü Görüntüle")
        print("2. Satın Alma")
        print("3. Çıkış")
        
        choice = input("Yapmak istediğiniz işlemi seçiniz (1/2/3): ")
        
        if choice == "1":
            display_menu()
        elif choice == "2":
            purchase_smoothie()
        elif choice == "3":
            print("Programdan çıkılıyor...")
            break
        else:
            print("Geçersiz seçenek. Lütfen 1, 2 veya 3'ü giriniz.")

if __name__ == "__main__":
    main()
```
