# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle:
    def __init__(self, name, speed):
        self.name = name
        self.speed = speed

    def display_info(self):
        return f"Vehicle Name: {self.name}, Speed: {self.speed} km/h"

class Bus(Vehicle):
    def __init__(self, name, speed, seating_capacity=50):
        super().__init__(name, speed)  
        self.seating_capacity = seating_capacity

    def seating_capacity_info(self):
        return f"The seating capacity of a bus is {self.seating_capacity} passengers"

bus = Bus("City Bus", 80)
print(bus.seating_capacity_info())
````

# Problem 2


School_bus'ın aynı zamanda Vehicle sınıfının bir örneği olup olmadığını belirleyiniz.

````python
class Vehicle:
    def __init__(self, name, speed):
        self.name = name
        self.speed = speed

    def display_info(self):
        return f"Vehicle Name: {self.name}, Speed: {self.speed} km/h"

class Bus(Vehicle):
    def __init__(self, name, speed, seating_capacity=50):
        super().__init__(name, speed)
        self.seating_capacity = seating_capacity

    def seating_capacity_info(self):
        return f"The seating capacity of a bus is {self.seating_capacity} passengers"

class SchoolBus(Bus):
    def __init__(self, name, speed, seating_capacity=50, school_name="Unknown"):
        super().__init__(name, speed, seating_capacity)
        self.school_name = school_name

    def school_info(self):
        return f"School Name: {self.school_name}"



school_bus = SchoolBus("School Bus", 60, 45, "Greenwood High")


if isinstance(school_bus, Vehicle):
    print("SchoolBus, Vehicle sınıfının bir örneğidir.")
else:
    print("SchoolBus, Vehicle sınıfının bir örneği değildir.")
````
