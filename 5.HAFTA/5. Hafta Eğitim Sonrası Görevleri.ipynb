# Problem 1

Fizikte, bir nesnenin sabit ivmeyle hareket ederken son hızını 
(veya hızını) bulmak için aşağıdaki denklem kullanılabilir:
```
vf = vi + at
burada:
vf= son hız
vi= ilk hız
a= hızlanma
t= zaman
```
İlk hız, ivme ve zaman verildiğinde, son hızı döndürecek bir fonksiyon yazın.

````python
def son_hiz_hesapla(vi, a, t):
    vf = vi + a * t
    return vf

vi = float(input("Başlangıç hızını giriniz (m/s): "))
a = float(input("İvme değerini giriniz (m/s^2): "))
t = float(input("Zamanı giriniz (s): "))

son_hiz = son_hiz_hesapla(vi, a, t)
print(f"Son hız: {son_hiz} m/s")
````

# Problem 2

1'den 1000'e kadar olan sayılardan mükemmel sayı olanları ekrana yazdırın. Bunun için bir sayının mükemmel olup olmadığını dönen bir tane fonksiyon yazın.

Bir sayının bölenlerinin toplamı kendine eşitse bu sayı mükemmel bir sayıdır. Örnek olarak 6 mükemmel bir sayıdır (1 + 2 + 3 = 6).

````python
def mukemmel_sayi_mi(sayi):
    toplam = 0
    for i in range(1, sayi):
        if sayi % i == 0:
            toplam += i
    return toplam == sayi

# 1'den 1000'e kadar mükemmel sayıları bulma ve ekrana yazdırma
for sayi in range(1, 1001):
    if mukemmel_sayi_mi(sayi):
        print(sayi)
````

# Problem 3

1'den 100'e kadar olan sayılardan pisagor üçgeni oluşturanları ekrana yazdıran bir fonksiyon yazın.(a <= 100,b <= 100)


````python
def pisagor_ucgenleri():
    for a in range(1, 101):
        for b in range(1, 101):
            c = (a**2 + b**2) ** 0.5  
            if c.is_integer() and c <= 100:  # c'nin bir tam sayı olup olmadığını ve 100'den küçük veya eşit olduğunu kontrol edecek
                print(f"Pisagor üçgeni bulundu: ({a}, {b}, {int(c)})")

pisagor_ucgenleri()
````

# Problem 4

Bir duvar boyamaya karar verdiniz. Boya kutusunun üzerindeki talimatta, 1 kutu boyanın 5 metrekarelik bir duvarı boyayabileceği yazıyor. Rastgele bir duvar yüksekliği ve genişliği verildiğinde, kaç kutu boya satın almanız gerektiğini hesaplayın.

kutu sayısı = (duvar yüksekliği * duvar genişliği) ÷ kutu başına kaplama.

örneğin Yükseklik = 2, Genişlik = 4, Kaplama = 5

kutu sayısı = (2 * 4) ÷ 5 = 1.6

Ancak bir kutu boyanın 0,6'sını satın alamayacağınız için, sonuç 2 kutuya yuvarlanmalıdır .


````python
import math
def boya_kutu_sayisi(yukseklik, genislik, kaplama=5):
    alan = yukseklik * genislik
    kutu_sayisi = math.ceil(alan / kaplama) # tam sayıya yuvarlama
    return kutu_sayisi

yukseklik = float(input("Duvarın yüksekliğini giriniz (metre): "))
genislik = float(input("Duvarın genişliğini giriniz (metre): "))

gerekli_kutu_sayisi = boya_kutu_sayisi(yukseklik, genislik)
print(f"Duvarı boyamak için gerekli kutu sayısı: {gerekli_kutu_sayisi}")
````

# Problem 5

Bir kelimedeki ünlü ve ünsüz harfleri sayan bir fonksiyon yazınız.

````python
def harf_sayaci(kelime):
    unlu_harfler = "aeıioöuü"
    unlu_sayisi = 0
    unsuz_sayisi = 0
    
    for harf in kelime.lower():  
        if harf.isalpha():  # Harf olup olmadığını kontrol ediyor
            if harf in unlu_harfler:
                unlu_sayisi += 1
            else:
                unsuz_sayisi += 1
    return unlu_sayisi, unsuz_sayisi

kelime = input("Bir kelime giriniz: ")
unlu_sayisi, unsuz_sayisi = harf_sayaci(kelime)
print(f"Ünlü harf sayısı: {unlu_sayisi}")
print(f"Ünsüz harf sayısı: {unsuz_sayisi}")
````

# Problem 6

Verilen bir listeden çift sayıları yazdıran bir fonksiyon yazınız.


````python
def cift_sayilari_yazdir(liste):
    for sayi in liste:
        if sayi % 2 == 0: 
            print(sayi)

sayi_listesi = list(map(int, input("Bir liste giriniz (virgülle belirtiniz): ").split(','))) #map() fonksiyonu, belirtilen bir işlemi (örneğin, bir dönüşüm) bir dizideki her öğeye uygulamak için kullanılır. 
print("Çift sayılar:")
cift_sayilari_yazdir(sayi_listesi)
````

# Problem 7

0'dan 10'a kadar olan sayıların toplamını hesaplayan özyinelemeli (Recursive Function) bir fonksiyon oluşturan bir fonksiyon yazınız.

Özyinelemeli (Recursive Function) bir işlev, kendini tekrar tekrar çağıran bir işlevdir.

````python
def toplam(n):
    if n == 0:  # Temel durum: n 0'a eşit olduğunda, toplam 0'dır
        return 0
    else:
        return n + toplam(n - 1)  # Özyineleme adımı: n ve n-1'in toplamı

# 0'dan 10'a kadar olan sayıların toplamını hesaplayalım
sonuc = toplam(10)
print(f"0'dan 10'a kadar olan sayıların toplamı: {sonuc}")
````
